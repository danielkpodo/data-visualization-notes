```python
import pandas as pd
from matplotlib import pyplot as plt

# fills are great for making our plots look professional but also they can give us some helpful insights on how to use them
# We can use conditional fields to show us what areas of our field fall far or below a certain threshold

data = pd.read_csv('fill_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')


plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
# You can also fill in b/n plts using the "dev_salaries" as median
# Lets say we want to fill the entire area underneath our python data
# alpha will help us see through it

# We also have a 3rd argument called y2 which tells us at what stage python salaries cross the median salary
# In the graph when our python salaries is below the overall salary it fills up and when above it fills down to that point
# lets say we want one fill with the color to green when we above median salary and another to red when below
# interpolate = true , makes sure certain x -intercessions do not get clip, and that all of the regions are filled correctly

plt.fill_between(ages,py_salaries,overall_median, where=(py_salaries > overall_median),
                 interpolate=True, alpha=0.25, label="Above Average")

# If you want to show the area below without a diff color then we need to write another fill between method
# We can pass in custom colors
plt.fill_between(ages,py_salaries,overall_median, where=(py_salaries <= overall_median),
                 interpolate=True, color='red', alpha=0.25, label="Below Average")


plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
