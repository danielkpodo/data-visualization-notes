```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates

plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

# for this use plt.plot_date()
# by default the dots are to be connected but we can fix that using linestyle=solid
plt.plot_date(dates, y, linestyle='solid')

# It is good to format the dates using autoformat figure si that they rotate our dates so that they fit a bit nicer
# gcf -> stands for get current figure
plt.gcf().autofmt_xdate()

# Lets seet how we can format the dates, b== Month, d== day, Y==year
date_format = mpl_dates.DateFormatter('%b, %d, %Y')

# Now grab the current axis to chage the dates, gca == get current axis
plt.gca().xaxis.set_major_formatter(date_format)

# data = pd.read_csv('data.csv')
# price_date = data['Date']
# price_close = data['Close']

# plt.title('Bitcoin Prices')
# plt.xlabel('Date')
# plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates

plt.style.use('Solarize_Light2')

data = pd.read_csv('time-series.csv')
# Matplotlib is reading the date field as string so you have to chage it
# To fix this we will use some pandas methods and later sort the values them
data["Date"] = pd.to_datetime(data['Date'])
data.sort_values("Date", inplace=True)


price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()



plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout();
```
