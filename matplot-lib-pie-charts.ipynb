## Basic Pie chart

```python
# basic pie chart plotting
slices = [140, 60, 200, 180, 300]

# You can add labels by using a list
labels = ["Apples", "Banana", "Tomatoes", "Oranges", "Cashew"]

# to add in some separtor color we pass in a wedge props -> wedgeprops={'edgeColor': "black"}

# To change the colors of the labels we can do the same thing we did for the labels using list in order
colors = ["red", "green", "cyan", "blue", "magenta"] # you can use hex values to get the color that u want

plt.rcParams["figure.figsize"] = (10,6)

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgeColor': 'black'});
plt.title("My Awesome Chart")
plt.tight_layout();
```

## Real Word use Case of Pie charts

```python
import matplotlib.pyplot as plt
# Note: Pie Charts do not look good when you have too much data, they are good for atmost five things
# Language Popularity,  for more values use horizontal bar
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode=[0, 0, 0, 0.1, 0] # the values represent how far it is from the radius

# Alot of the time you will see some pie charts that have emphasis on some parts of the chart, we achieve this with explode
# ...argument that will offset the slice, we set up an explode list

# We can also add asethetic to our chart by passing an argument shadow=True, which gives you 3D feel

# We can also change the starting angle of our pie, where it should start from -> startangl=90

# If we want to display percentages that each value is taking then we could do that as well -> autopct='%1.1f%%'


plt.title("Popular Languages in 2020")
plt.tight_layout()


plt.pie(slices, labels=labels, wedgeprops={'edgeColor': "black"},
        explode=explode, shadow=True, autopct='%1.1f%%', startangle=90);
```
