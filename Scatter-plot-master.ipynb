# Working with Scatter Plots

+ Scatter plots are great when you want to show the relationship between values

```python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

# You can pass in the sizes list for the size, this is essesntially good for heart disease analysis tracking obesity
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
       538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

#to plot we use 
# scatter plots are great for seeing different trends or outliers
# since our plots looks scattered like this it tells us that there is no correlation b/n our two list of values

# to change the size of the dots if it looks small we use s=100
# we can also chage the color of the dots using c="colorname or hexcode"
# Scatter also looks alot nicer if we add edges to the circle and add a bit of alpha values so we can see through them
# ... a bit

# lets pretend that this people rate something from 1 - 10 and we want to somehow plot their rating as well, well to do 
# ...that we can assign diff numbers for diff possibilities

# You can also change color of plot based of these rating, each of these values wil correspond to a data point
# in our x and y values, you passed this to the c attribute in the plot
rating = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

# Now if you do not like the colors in the plot you can chage them using a color map, check docs, pass cmpa='value'
# Now the shades are dependent based of the ratings, NB, the lighter ones are close to zero

# It is also good to add a label to our colormap so people viewing our chart knows what this colors represent
# To do this we use cbar=plt.colorbar() next -> cbar.set_label("Ratings of Customers")

# I replace s=100, with the sizes list
plt.scatter(x, y, s=sizes,c=rating,cmap='Oranges',edgecolor="black",linewidth=1, alpha=0.75);

# This should be under the plot else it wont work
cbar = plt.colorbar()
cbar.set_label("Satisfaction")
```

```python
# From the graph you can deduce that the dark colored ones are very satisfied
```

# Plotting Real World Data from a CSV file

```python
plt.style.use('dark_background')

# This is to check if there is a correlation b/n the view count and the likes
# So we can take a guess that says the more views a video has the more likes it is going to get
data = pd.read_csv('scatter.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.tight_layout()
plt.grid(False)


plt.scatter(view_count, likes,edgecolor="black", c=ratio, linewidth=1, alpha=0.75, cmap='summer');

cbar = plt.colorbar()
cbar.set_label("Like / Dislike Ratio")

# Now note when you have an outlier it also good to use a log scale lessen how much those outliers skew the plot
plt.xscale('log')
plt.yscale('log')
```

```python
# In the plot there is one person with more views and likes accounting as an outlier
# in determining the like / dislike the ones more in yellow perform better and those in green perform poorly
```
