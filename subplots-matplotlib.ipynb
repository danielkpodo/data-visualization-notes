```python
# To work with your plots in a more objected-oriented manner then it is best to use the subplot method
# A lot of people prefer to make their plots this way even it is a single plot
# Using this can help us make multiple plots, it is useful when working with multiple figures and axes
```

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

data = pd.read_csv('subplots.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# using the objected oriented approach

fig, ax = plt.subplots()


ax.plot(ages, py_salaries, label='Python')
ax.plot(ages, js_salaries, label='JavaScript')

ax.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax.legend()

ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()

# Why is this approach useful ?
# It can help us divide our plots into multiple plots instead of one plot
```

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

data = pd.read_csv('subplots.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# using the objected oriented approach

# Why is this approach useful ?
# It can help us divide our plots into multiple plots instead of one plot
# It allows us to set the nrows and ncols

# To ensure that we use only one x-axis we can pass an argument known as sharex
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')



ax.legend()
ax1.set_title('Median Salary (USD) by Age')
# Remove xlabel on this axis
#ax1.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

ax2.legend()
# since the two titles are the same I can remove ax2 title
# ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()

## So now we can see that we have one figure with two different plots
# Thses plot is on the same figure
```
